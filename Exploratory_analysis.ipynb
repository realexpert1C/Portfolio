```python
for r in data['rooms'].unique():
    data.loc[(data['rooms'] == r) & (data['living_area'].isna()), 'living_area'] =\
    data.groupby('rooms').median()['living_area'][r]
```

```python
for r in data['rooms'].unique():
    data.loc[(data['rooms'] == r) & (data['kitchen_area'].isna()), 'kitchen_area'] =\
    data.groupby('rooms').median()['kitchen_area'][r]
data.loc[(data['rooms'] == 0) & (data['kitchen_area'].isna()),
          'kitchen_area'] = data.fillna(0)
```

```python
def check_area(x):
    k = x.loc['kitchen_area']
    l = x.loc['living_area']
    t = x.loc['total_area']
    # print('start', k)
    if (k + l) > t:
        k = 0.8 * k
        l = 0.8 * l
        x.loc['kitchen_area'] = k
        x.loc['living_area'] = l
    # print('finish', k)
    return x
```

```python
data = data.apply(check_area, axis=1)
```

def search_out(column, coef=1.5, share_out_limit=0.1):
    p25 = np.quantile(column, 0.25)
    p75 = np.quantile(column, 0.75)

    return (1 - (
                (column >= p25 - coef*(p75-p25)) & 
                (column <= p75 + coef*(p75-p25))
            ).mean()) > share_out_limit
